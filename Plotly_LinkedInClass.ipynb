# Scratter Plot with Mapbox and  Geo Pandas

```python
!pip install geopandas
import geopandas as gdp
import pandas as pd
```

```python
peaks = pd.read_csv('co_14ers.csv')
peaks = gdp.GeoDataFrame(peaks, geometry=gdp.points_from_xy(peaks.lon, peaks.lat))
peaks.head()
```

```python
peaks['size'] = 15
fig = px.scatter_mapbox(peaks,
                        lat = peaks.geometry.y,
                        lon = peaks.geometry.x,
                        color = "Range",
                        hover_name = "Peak",
                        size = "size",
                        zoom = 5 )
```

```python
fig.update_layout(
    title = '14ers of colorado',
    mapbox_style = 'white-bg',
    mapbox_layers = [
        {
            "below": "traces",
            "sourcetype" : "raster",
            "sourceattribution": "United States Geological Survey",
            "source" : [
                "https://basemap.nationalmap.gov/arcgis/rest/services/USGSImageryOnly/MapServer/tile/{z}/{y}/{x}"
            ]
        }
    ])
fig.show()
```

```python
#sunburst chart
df_sample_super_store= pd.read_excel('Sample - Superstore.xls',header = 0, sheet_name = "Orders")
df_sample_super_store.head(10)
```

```python
grouped_segment = df_sample_super_store.groupby('Segment').size().reset_index(name='count')
print(grouped_segment)

grouped_cat= df_sample_super_store.groupby('Category').size().reset_index(name='count')
print(grouped_cat)

Salesby_segment_cat = df_sample_super_store.groupby(['Segment','Category']).size().reset_index(name='count')
print(Salesby_segment_cat)
```

```python
fig = px.sunburst(
    df_sample_super_store,
    path = ["Category", "Sub-Category"],
    values = "Sales",
    branchvalues = "total"
)
fig.show()
```

```python
import plotly.graph_objects as go

nodes =  ["Corporate", "Consumer", "Home Office","Technology", "Furniture", "Office Supplies"]
link_source = Salesby_segment_cat['Segment']
link_target =  Salesby_segment_cat['Category']
link_value = Salesby_segment_cat['count']
print(link_source)

sankey_trace = go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=nodes,
    ),
    link=dict(
        source=link_source,
        target=link_target,
        value=link_value,
    )
)

layout = go.Layout(title="Sankey Diagram")

fig = go.Figure(data=[sankey_trace], layout=layout)
fig.show()
```
